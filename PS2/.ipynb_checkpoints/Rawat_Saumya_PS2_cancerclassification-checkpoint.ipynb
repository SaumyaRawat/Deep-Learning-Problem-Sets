```python
import numpy as np
from numpy import genfromtxt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

import keras 
from keras.wrappers.scikit_learn import KerasClassifier
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.models import load_model

import matplotlib.pyplot as plt

```

## Load Data

```python
dataset = genfromtxt('Breast Cancer/breastCancerData.csv', delimiter=',')
y = genfromtxt('Breast Cancer/breastCancerLabels.csv', delimiter=',')


# Perform Train, Test split
x_train, x_test, y_train, y_test = train_test_split(dataset, y, test_size = 0.15, random_state = 0)
```

## Pre-process data
1. Standardize dataset
2. Ensure shape is compatible with required format for tensorflow:  (batch, height, width, channels)


```python
# Normalize
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)

print("Size of training set: ",x_train.shape)
print("Size of test set: ",x_test.shape)


#print("Some examples of the training set:")

#for i in range(0, 3):
#    plt.subplot(330 + (i+1))
#    plt.imshow(x_train[i].reshape(28,28), cmap=plt.get_cmap('gray'))
```

## Define the ANN

Let the network be defined with a single layer having ReLu activation and a sigmoid output layer due to the presence of 2 classes. 



```python
# Initialising the ANN

model = Sequential([
    Dense(units = 16, kernel_initializer = 'uniform', activation = 'relu', input_dim = x_train.shape[1]),
    Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid')
])

# Initialize and compile:
keras.initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None)
model.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.summary()
```

## Train model
Batch Size: 5
Epochs: 100
Validation Size: 20% 


```python
# Training
batch_size = 5
history = model.fit(
    x_train, 
    y_train, 
    validation_split=0.2,
    batch_size = batch_size,
    epochs = 100
)
#model.save('breast_cancer_model.h5')
```

```python
score = model.evaluate(x_test,y_test,verbose=0)

print('Test Loss : {:.4f}'.format(score[0]))
print('Test Accuracy : {:.4f}'.format(score[1]))

```

## Visualizing the performance of the networks

```python


accuracy = history.history['acc']
val_accuracy = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'g', label='Training Accuracy')

plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('epoch')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'g', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('epoch')

plt.legend()
plt.show()
```

The graphs indicate the network is not performing correctly on the data. We can see that underfitting has occured. We should add more layers to the model.

## Updating the model
Increasing the number of layers in the network. Both having ReLu activation and a sigmoid output layer due to the presence of 2 classes. 

Batch Size: 5
Epochs: 100
Validation Size: 20% 



```python
# Initialising the ANN

model2 = Sequential([
    Dense(16, activation='relu', input_dim = x_train.shape[1]),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])


# Initialize and compile:
keras.initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None)
model2.compile(optimizer=keras.optimizers.Adam(lr=0.001), loss=keras.losses.mean_squared_logarithmic_error, metrics=['accuracy'])
model2.summary()

```

```python
history = model2.fit( 
    x_train, 
    y_train, 
    batch_size=5, 
    epochs=100, 
    verbose=1, 
    validation_split=0.2
)
```

## Improved Accuracy

```python
score = model2.evaluate(x_test,y_test,verbose=0)


print('Test Loss : {:.4f}'.format(score[0]))
print('Test Accuracy : {:.4f}'.format(score[1]))
```

## Visualizing the updated results

```python


accuracy = history.history['acc']
val_accuracy = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'g', label='Training Accuracy')

plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('epoch')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'g', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('epoch')

plt.legend()
plt.show()

```

Since now training accuracy >> validation accuracy, overfitting seems to have occured. Adding more layers with a dropout layer should improve the results.


```python
model3 = Sequential([
    Dense(16, activation='relu', input_dim = x_train.shape[1]),
    Dense(8, activation='relu'),
    Dense(6, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Initialize and compile:
keras.initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None)
model3.compile(optimizer=keras.optimizers.Adam(lr = 0.0001), loss=keras.losses.mean_squared_logarithmic_error, metrics=['accuracy'])
model3.summary()

```

```python
history = model3.fit( 
    x_train, 
    y_train, 
    batch_size=1, 
    epochs=150, 
    verbose=1, 
    validation_split=0.2
    
)
```

## Final Accuracy

```python
score = model3.evaluate(x_test,y_test,verbose=0)

print('Test Loss : {:.4f}'.format(score[0]))
print('Test Accuracy : {:.4f}'.format(score[1]))
```

## Visualizing the final model accuracy and loss

```python
accuracy = history.history['acc']
val_accuracy = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))

plt.plot(epochs, accuracy, 'g', label='Training Accuracy')

plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('epoch')
plt.legend()
plt.figure()

plt.plot(epochs, loss, 'g', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('epoch')

plt.legend()
plt.show()

```
