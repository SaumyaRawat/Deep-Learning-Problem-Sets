```python
class ConvClassifier(nn.Module):
    def __init__(self):
        super(ConvClassifier, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(16)
        self.conv2 = nn.Conv2d(16, 16, kernel_size=3, padding=1, stride=1)
        self.bn2 = nn.BatchNorm2d(16)
        self.conv3 = nn.Conv2d(16, 32, kernel_size=3, padding=1, stride=1)
        self.bn3 = nn.BatchNorm2d(32)

        self.conv4 = nn.Conv2d(32, 16, kernel_size=3, padding=1, stride=1)
        self.bn4 = nn.BatchNorm2d(16)

        self.conv5 = nn.Conv2d(16, 16, kernel_size=5, padding=1, stride=2)
        self.bn5 = nn.BatchNorm2d(16)
        self.fc1 = nn.Linear(16*15*15, 100)
        self.fc2 = nn.Linear(100, 10)

        
    def forward(self, x):
        self.conv_l1 = self.conv1(x)
        self.conv_l1 = self.bn1(self.conv_l1)
        
        self.conv_l2 = F.relu(self.conv2(self.conv_l1))
        self.conv_l2 = self.bn2(self.conv_l2)
        
        self.conv_l3 = F.relu(self.conv3(self.conv_l2))
        self.conv_l3 = self.bn3(self.conv_l3)

        self.conv_l4 = F.relu(self.conv4(self.conv_l3))
        self.conv_l4 = self.bn4(self.conv_l4)

        self.conv_l5 = F.relu(self.conv5(self.conv_l4))
        self.conv_l5 = self.bn5(self.conv_l5)

        self.fc_l1 = self.conv_l5.view(-1, 16 * 15 * 15)
        self.fc_l1 = F.relu(self.fc1(self.fc_l1))
        self.fc_l2 = self.fc2(self.fc_l1) 
        
        return F.log_softmax(self.fc_l2, dim=1)
    
            
```

```python
img_loader = ImgLoader(X_train, y_train, None)
train_loader = DataLoader(img_loader, batch_size=5, shuffle=True, num_workers=0)
```

```python
def eval(model):
    model.train(False)
    count = 0
    for x,y in train_loader:
        #x = torch.from_numpy(np.array([x])).permute(0, 3, 1, 2).type(torch.LongTensor)        
        if use_gpu:
            x = x.to(cuda0)

        out = model(Variable(x).float())
        label = np.argmax(out.data.cpu().numpy()) # needs to be optimized
        
        print("We got: ",label ,"We wanted: ",y)
        #if Variable(y.squeeze(1)) == label:
        #    count += 1

    print(count * 1.0 / len(X_test))
```

```python
epochs = 30
learning_rate = 0.0005
criterion = nn.CrossEntropyLoss()
net = ConvClassifier()

if use_gpu:
    net = net.to(cuda0)

optimizer = optim.SGD(params=net.parameters(), lr=learning_rate)

for epoch in range(epochs):
    losses = []
    net.train(True)
    for x, y in train_loader:
        if use_gpu:
            x, y = x.to(cuda0), y.to(cuda0)
        optimizer.zero_grad()
        inputs =  Variable(x)
        inputs = inputs.float()
        output = net(inputs)
        #output = net(x)
        targets = Variable(y.squeeze(1))
        #targets = Variable(y)
        loss = criterion(output, targets)
        loss.backward()
        optimizer.step()
        losses.append(loss.item())

    if (epoch+1) % 1 == 0:
        print('Epoch %d Loss %.4f  ' % (epoch+1, np.average(losses)))
        print('test accuracy')
        eval(net)

```
