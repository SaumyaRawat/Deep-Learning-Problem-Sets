```python
import torch 
import torch.nn as nn
import torchvision.datasets as dsets
from skimage import transform
import torchvision.transforms as transforms
from torch.autograd import Variable
import pandas as pd
import numpy as np
from torch.utils.data import Dataset, DataLoader
import random
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import torch.optim as F

```

# Dataset introduction and loading

### The flowers dataset consists of 4323 32X32X3 images that need to be classified into 5 classes:  

1. daisy
2. dandelion
3. rose
4. sunflower
5. tulip.  

### The dataset is split into 0.85% Train and 0.15% Test set.

### Data augmentation:
* Since the dataset is relatively small, we perform some data augmentations to ensure model robustness and improve test accuracy.
* The test accuracy without data augmentation was: 69.5%
* The test accuracy after data augmentation is: 81%
* The following transforms are applied:
    1. transforms.RandomCrop(): Crops from image randomly 
    2. transforms.RandomHorizontalFlip():  With a default probability of 0.5%, Randomly flips the image horizontally\
    3. Normalize:  Since ToTensor() converts the PIL Image from range [0,255]  to a FloatTensor of shape (C x H x W) with range [0.0, 1.0] we have to renormalize the input to [-1, 1] based on the following formula,
    \begin{equation}
    input = \frac{input−μ}{\sigma}
    \end{equation}

```python
# Load Data
img_data = np.load("Flowers/flower_imgs.npy") #(4323, 32, 32, 3)
labels = np.load("Flowers/flower_labels.npy")

num_classes=max(labels)
imgs = img_data
print("Dataset dimension: ",imgs.shape)

img_mean = np.mean(np.swapaxes(imgs/255.0,0,1).reshape(3, -1), 1)
img_std = np.std(np.swapaxes(imgs/255.0,0,1).reshape(3, -1), 1)
print("mean: {}, std: {}".format(img_mean, img_std))

# Train - test split
X_train, X_test, y_train, y_test = train_test_split(imgs,labels, test_size = 0.15, random_state = 0)


class FlowerDataset(Dataset):
    '''Flowers Dataset'''
    def __init__(self, imgs, labels, transform=None):
        """
        Args:
            transform (callable): Optional transform to apply to sample
        """
        self.X = imgs
        self.Y = labels
        self.transform = transform
        
    def __len__(self):
        return len(self.X)
    
    def __getitem__(self, idx):
        item = self.X[idx]
        label = self.Y[idx]
        
        if self.transform:
            item = self.transform(item)
        
        return (item, label)

train_transform, test_transform = None, None
normalize = transforms.Normalize(mean=list(img_mean),
                                 std=list(img_std))

train_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.RandomCrop(32, padding=4),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        normalize,
])
test_transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.ToTensor(),
        normalize,
])

train_dataset = FlowerDataset(X_train, y_train, train_transform)
test_dataset = FlowerDataset(X_test, y_test, test_transform)

labels_map = {0 : 'Daisies', 1 : 'Dandelion', 2 : 'Rose', 3 : 'Sunflower', 4 : 'Tulip'}
w=10
h=10
fig=plt.figure(figsize=(8, 8))
columns = 4
rows = 5
for t in range(1, columns*rows +1):
    i = random.randint(0, len(imgs))
    fig.add_subplot(rows, columns, t)
    plt.title(labels_map[labels[i]])
    plt.imshow(img_data[i])
plt.show()
```

# The model
Since this is an image classification task, a VGG model is chosen as it is a well known network with good accuracies reported on the ImageNet task. VGG 11 consists of 11 layers with:
1. Convolutions layers
2. Max pooling layers
3. Fully connected layers at end


```python
# Network Parameters
num_epochs = 100
batch_size = 70
learning_rate = 10e-3
momentum=0.9
weight_decay=5e-4

if torch.cuda.is_available():
    print("Using CUDA")
    
```

```python
nThreads = 4
cuda0 = torch.device('cuda:0')  # CUDA GPU 0


cfg = {
    #'VGG11': [64, 'M', 128, 'M', 256, 256, 'M', 512, 512, 'M', 512, 512, 'M'],
    'VGG11': [64, 'M', 128, 'M', 256, 256, 'M', 512, 512, 'M', 512, 512, 'M'],
    'VGG13': [64, 64, 'M', 128, 128, 'M', 256, 256, 'M', 512, 512, 'M', 512, 512, 'M'],
    'VGG16': [64, 64, 'M', 128, 128, 'M', 256, 256, 256, 'M', 512, 512, 512, 'M', 512, 512, 512, 'M'],
    'VGG19': [64, 64, 'M', 128, 128, 'M', 256, 256, 256, 256, 'M', 512, 512, 512, 512, 'M', 512, 512, 512, 512, 'M'],
}


class VGG(nn.Module):
    def __init__(self, vgg_name):
        super(VGG, self).__init__()
        self.features = self._make_layers(cfg[vgg_name])
        #self.classifier = nn.Linear(512, 10)
        self.classifier = nn.Linear(512, 5)
        
    def forward(self, x):
        out = self.features(x)
        out = out.view(out.size(0), -1)
        out = self.classifier(out)
        return out

    def _make_layers(self, cfg):
        layers = []
        in_channels = 3
        for x in cfg:
            if x == 'M':
                layers += [nn.MaxPool2d(kernel_size=2, stride=2)]
                #layers += [nn.MaxPool2d(kernel_size=2, stride=2, dilation=1)]                
            else:
                layers += [nn.Conv2d(in_channels, x, kernel_size=3, padding=1),
                           nn.BatchNorm2d(x),
                           nn.ReLU(inplace=True)]
                in_channels = x
        layers += [nn.AvgPool2d(kernel_size=1, stride=1)]
        return nn.Sequential(*layers)
```

## Loss function and optimizer
Loss function used: Cross Entropy Loss
Optimizer: An SGD optimizer  with 
1. num_epochs = 100
2. batch_size = 70
3. learning_rate = 10e-3
4. momentum=0.9
5. weight_decay=5e-4

Step decay, a form of learning rate annealing is also applied. At the completion of *50%* and *75%* of training, the learning rate is multiplied by gamma=0.1. We'd like to traverse the initial parameters quickly to a range of "good" parameter values but to explore the deeper & narrower parts of the loss function at a slower pace with a small learning rate


```python
#instance of the Conv Net
cnn = VGG('VGG13')
if torch.cuda.is_available():
    print("Using CUDA")
    cnn.to(cuda0)

#loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(cnn.parameters(), lr=learning_rate, momentum=momentum, weight_decay=weight_decay)
scheduler = F.lr_scheduler.MultiStepLR(optimizer, milestones=[int(0.5*num_epochs),int(0.75*num_epochs)], gamma=0.1)
```

# Training the Model

```python
train_losses = []
train_accuracy = []

for epoch in range(num_epochs):
    scheduler.step()
    train_acc = 0
    train_loss = 0
    t = 0
    train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                batch_size=batch_size,
                shuffle=True)
    for i, (images, labels) in enumerate(train_loader):
        if torch.cuda.is_available():
            images = images.to(cuda0)
            labels = labels.to(cuda0)
        images = Variable(images.float())
        labels = Variable(labels)
        
        # Forward + Backward + Optimize
        optimizer.zero_grad()
        outputs = cnn(images)
        
        # LOSS
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
        
        # ACCURACY
        _, predicted = torch.max(outputs.data, 1)
        pred_y = predicted.data.cpu().numpy().squeeze()
        target_y = labels.cpu().data.numpy()
        train_acc += (float((pred_y == target_y).astype(int).sum()) / float(target_y.size))
        t+=1
    train_accuracy.append(train_acc/t)
    train_losses.append(train_loss/t)

    if (epoch) % 10 == 0:
        print ('Epoch : %d/%d, Loss: %.4f' %(epoch+1, num_epochs, (train_loss/t)))
```

# Testing the model

## Final test accuracy is 80.9753 %


```python
cnn.eval()
test_losses = []
test_accuracy = []

for epoch in range(num_epochs):
    test_loader = torch.utils.data.DataLoader(dataset=test_dataset,
                                              batch_size=batch_size,
                                              shuffle=True)

    epoch_acc = 0
    epoch_loss = 0
    c = 0
    for images, labels in test_loader:
        if torch.cuda.is_available():
            images = images.to(cuda0)
            labels = labels.to(cuda0)
        images = Variable(images.float())
        labels = Variable(labels)

        optimizer.zero_grad()
        outputs = cnn(images)

        # LOSS
        loss = criterion(outputs, labels)
        epoch_loss += loss.item()

        # ACCURACY
        _, predicted = torch.max(outputs.data, 1)
        pred_y = predicted.data.cpu().numpy().squeeze()
        target_y = labels.cpu().data.numpy()
        epoch_acc += float((pred_y == target_y).astype(int).sum()) / float(target_y.size)
        
        c+=1
    
    test_accuracy.append(epoch_acc/c)
    test_losses.append(epoch_loss/c)

print('Test Accuracy of the model on test images: %.4f %%' % (100 * (sum(test_accuracy) / num_epochs)))

```

```python
epochs = range(num_epochs)

plt.plot(epochs, train_accuracy, 'g', label='Training Accuracy')
plt.plot(epochs, test_accuracy, 'b', label='Validation Accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('epoch')
plt.legend()
plt.figure()

plt.plot(epochs, train_losses, 'g', label='Training Loss')
plt.plot(epochs, test_losses, 'b', label='Validation Loss')
plt.title('Training and validation loss')
plt.xlabel('epoch')
```
